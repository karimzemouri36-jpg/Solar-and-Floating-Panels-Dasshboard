```python
import random
import matplotlib.pyplot as plt

# Nombre de simulations
X = 100

# Liste pour stocker les sommes
sommes = []

# Simulation des lancers
for i in range(X):
    A= random.randint(1, 6)
    B = random.randint(1, 6)
    S = A + B
    sommes.append(S)

# Histogramme
plt.hist(sommes, bins=range(2, 14), align='left', edgecolor='red')
plt.xlabel("SSomme des deux dés")
plt.ylabel("Y name")
plt.title(f"Histogramme des sommes pour {X} lancers de deux dés")
plt.xticks(range(2, 13))
plt.show()
```
