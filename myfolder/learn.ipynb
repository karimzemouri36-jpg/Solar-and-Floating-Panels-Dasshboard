```python
import pandas as pd

df1=[("Karim,23,Algeria,21367535,23/05/1990"),
      ("Said,30,Morocco,21367536,15/08/1985"),
      ("Amina,27,Tunisia,10/12/1992"),]
df2=[("Karim", "Engineer"),
      ("Said", "Doctor"),   
      ("Amina", "Artist"),]
x=pd.DataFrame([s.split(',') for s in df1], columns=["Name", "Age", "Country", "ID", "DOB"])
y=pd.DataFrame(df2,columns=["Name","Profession"])
print(x)
print(y)
z=pd.merge(x,y, on="Name")
print(z)
```

```python
df1=[("Karim", 3000),
      ("Said", 4000),
      ("Amina", 5000),
    ("Rami", 3500),
      ("Lina", 3500)]
df2=[("Karim", "Engineer"),
      ("Said", "Engineer"),
        ("Amina", "Artist"),
        ("Rami", "Teacher"),
        ("Lina", "Artist"),]
x=pd.DataFrame(df1,columns=["Name","Salary"])
y=pd.DataFrame(df2,columns=["Name","Profession"])
z=pd.merge(x,y, on="Name")
print(z)
g1=z.groupby("Profession").count()
g2=z.groupby("Profession").sum()
g3=z.groupby("Profession")["Salary"].mean()
print(g1)
print(g2)
print(g3)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np  
plt.style.use('dark_background')
x=np.array(['Sanday','Monday','Tuesday','Wednesday','Thursday'])
y=np.array([2,3,5,7,11])
z=np.array([1,4,9,16,25])
plt.plot(x,y, color='red', marker='o', linestyle='-', linewidth=1, label='2024')
plt.plot(x,z, color='blue', marker='s', linestyle='--', linewidth=1 , label='2025')
plt.bar(x, y, color='red', edgecolor='black', alpha=0.7, label='2024')
plt.bar(x, z, color='blue', edgecolor='black', alpha=0.7, label='2025')

plt.title('Performane of our campany')
plt.xlabel('Months')
plt.ylabel('Production')
plt.legend()
plt.show()
plt.pie(z, labels=x, autopct='%1.1f%%', startangle=140, colors=['red', 'blue', 'green', 'yellow', 'purple']     )
plt.show()
```
